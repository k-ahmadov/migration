```python
import numpy as np
import matplotlib.pyplot as plt
import similarity_diffusion_solver as sds
import pickle

plt.rcParams.update({
    'font.family': 'serif',
    'font.serif': 'Times',
    'text.usetex': True,
    'text.latex.preamble': r'\usepackage{amsmath}',
})
```

```python
with open('properties_q.pkl', 'rb') as f:
    props = pickle.load(f)

with open('results_q.pkl', 'rb') as f:
    results = pickle.load(f)
```

```python
MU =  props['Fluid Viscosity']
KN_VALUES = tuple(results.keys())
P_x_t = {kn_value: results[kn_value]['P-x'] for kn_value in KN_VALUES}
w_x_t = {kn_value: results[kn_value]['w-x'] for kn_value in KN_VALUES}
wi = props['Initial Aperture']
q = props['Injection Rate']
```

```python
kn = 10e9

row_indices = []
i = 100
while i < len(w_x_t[kn]) - 15000:
    row_indices.append(i)
    i += int(100 + 0.1 * i)  # Example: linearly increasing step size
len(row_indices)
```

```python
kn = 10e9
a = kn / (12 * MU)
x_3dec = w_x_t[kn].iloc[0, 1:].astype(float).to_numpy()

fig, axs = plt.subplots(1, 2, figsize=(5.5, 2.8), dpi=300)
cmap = plt.cm.Greys  # Grayscale colormap
row_indices = range(500, len(w_x_t[kn]), 500)
n_steps = len(row_indices)

for i, row_idx in enumerate(row_indices):
    t = float(w_x_t[kn].iloc[row_idx, 0])
    w_x_3dec = w_x_t[kn].iloc[row_idx, 1:].astype(float).to_numpy()
    # P_x_3dec = P_x_t[kn].iloc[row_idx, 1:].astype(float).to_numpy()
    # w_x_3dec = wi + P_x_3dec / kn

    # Normalize color: early = light gray, late = dark
    color = cmap(i / n_steps + 0.1)

    axs[0].plot(x_3dec, w_x_3dec * 1e3, color=color)

    # Nondimensionalize
    zeta_3dec = x_3dec / (a * q**3 * t**4)**(1/5)
    theta_3dec = w_x_3dec / ((q**2 * t) / a)**(1/5)
    axs[1].plot(zeta_3dec, theta_3dec, color=color)

axs[0].set(
    xlabel=r'Distance in the fracture [m], $x$',
    ylabel=r'Fracture aperture [mm], $w$',
)

axs[1].set(
    xlabel=r'Dimensionless distance, $\zeta = \frac{x}{\left(a q^3 t^4\right)^{1/5}}$',
    ylabel=r'Dimensionless aperture, $\theta = \frac{w}{\left(\frac{q^2 t}{a}\right)^{1/5}}$',
    xlim=(0, 5),
)

for ax in axs:
    ax.grid(True, alpha=0.1)

plt.suptitle(fr'$k_n = {kn/1e9}$ GPa/m')
plt.tight_layout()
plt.show()
```
