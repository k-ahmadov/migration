```python
import numpy as np
import matplotlib.pyplot as plt
import similarity_diffusion_solver as sds
import pickle
```

```python
with open('properties.pkl', 'rb') as f:
    props = pickle.load(f)

with open('results.pkl', 'rb') as f:
    results = pickle.load(f)
```

```python
KN_VALUES = tuple(results.keys())
MU =  props['Fluid Viscosity']
P_x_t = {kn_value: results[kn_value]['P-x'] for kn_value in KN_VALUES}
wi = props['Initial Aperture']
w0_ = {kn_value: props[kn_value]['Aperture at x=0'] for kn_value in KN_VALUES}
times_ = {10e9: [0.1, 1, 4, 7.5], 30e9: [1, 5, 15, 30], 2000e9: [0.1, 1, 2, 3]}
```

```python
solutions = {}
n_values = ("n=0", "n=3")

for kn_value in KN_VALUES:
    solutions[kn_value] = {}
    for n in n_values:  # Iterate over the n_values tuple
        if n == "n=0":
            solutions[kn_value][n] = sds.solve_exact_similarity_n0(wi / w0_[kn_value])
        elif n == "n=3":
            solutions[kn_value][n] = sds.solve_n3(wi / w0_[kn_value])
```

```python
kn = 30e9
x_3dec = P_x_t[kn].iloc[0, 1:].astype(float).to_numpy()

fig, ax = plt.subplots(figsize=(3.3, 2.7), dpi=300,layout='constrained')

for t in times_[kn]:
    c = 0.7 * t / max(times_[kn]) + 0.3
    row_index = sds.findRowIndexWithSpecificTime(P_x_t[kn], t)
    P_x_3dec = P_x_t[kn].iloc[row_index, 1:].astype(float).to_numpy()
    ax.plot(x_3dec, P_x_3dec / 1e6, label=fr"3DEC,\ $t={t}\ $s", marker='^', linestyle='', color=plt.cm.Greys(c), markerfacecolor='none', alpha=0.3)  # Add linestyle here

    D = (kn * w0_[kn] ** 3) / (12 * MU)
    x, w = sds.dimensionalize(solutions[kn]['n=3'][0], solutions[kn]['n=3'][1], t, D, w0_[kn])
    P = kn * (w - wi)
    ax.plot(x, P / 1e6, label=fr"$n=3,\ t={t}\ $s", linestyle='-', color=plt.cm.Greys(c))  # Add linestyle here

ax.set_xlabel(r"Distance along the fracture $x$, [m]")
ax.set_ylabel(r"Fluid pressure, $P$ [MPa]")
ax.set_title(fr"$k_n={kn / 1e9}$ GPa")
ax.legend(fontsize=7, loc='best', title="Method / Time", frameon=False)
plt.show()
```

```python
marker_ = {10e9: 'D', 30e9: '^', 2000e9: 'o'}
x_3dec = P_x_t[kn].iloc[0, 1:].astype(float).to_numpy()
fig, ax = plt.subplots(figsize=(3.3, 2.7), dpi=300, layout='constrained')
for kn in KN_VALUES[:-1]:
    for row_index in range(2000, P_x_t[kn].shape[0], 2000):
        t = float(P_x_t[kn].iloc[row_index, 0])
        P_x_3dec = P_x_t[kn].iloc[row_index, 1:].astype(float).to_numpy()
        D = (kn * w0_[kn] ** 3) / (12 * MU)
        theta_3dec = (P_x_3dec / kn + wi) / w0_[kn]
        zeta_3dec = x_3dec / np.sqrt(D * t)
        ax.plot(zeta_3dec, theta_3dec, label=fr"3DEC,\ $t={t}\ $s", marker=marker_[kn], linestyle='', markerfacecolor='none', alpha=0.5, markersize=4, markevery=30)  
    zeta, theta = solutions[kn]['n=3'][0], solutions[kn]['n=3'][1]
    ax.plot(zeta, theta, linestyle='-', color='k')

ax.set_xlabel(r"Similarity variable $\zeta$")
ax.set_ylabel(r"Similarity variable $\theta$")
ax.set_xlim(-0.1, 3)
```
