```python
import matplotlib.pyplot as plt
import numpy as np
import similarity_diffusion_solver as sds
import pickle

plt.rcParams.update({
    'font.family': 'serif',
    'font.serif': 'Times',
    'text.usetex': True,
    'text.latex.preamble': r'\usepackage{amsmath}',
})
```

```python
with open('properties.pkl', 'rb') as f:
    props = pickle.load(f)

with open('results.pkl', 'rb') as f:
    results = pickle.load(f)
```

```python
MU =  props['Fluid Viscosity']
KN_VALUES = tuple(results.keys())
P_x_t = {kn_value: results[kn_value]['P-x'] for kn_value in KN_VALUES}
wi = props['Initial Aperture']
w0_ = {kn_value: props[kn_value]['Aperture at x=0'] for kn_value in KN_VALUES}
```

```python
kn = 2000e9

plt.figure(figsize=(3.3, 2.5), dpi=300)
zeta, theta = sds.solve_exact_similarity_n0(wi / w0_[kn])
overpressure_threshold = 1e6
zeta_interp = np.linspace(0, 4, 3000)
theta_interp = np.interp(zeta_interp, zeta, theta)
plt.plot(zeta_interp, theta_interp, '-', label='Similarity solution', color='black')
theta_threshold = (overpressure_threshold / kn + wi) / w0_[kn]
threshold_indices = np.where(theta_interp > theta_threshold)[0]
if len(threshold_indices) > 0:
    index_threshold = threshold_indices[-1]
    zeta_threshold = zeta_interp[index_threshold]
    plt.plot(zeta_threshold, theta_threshold, marker='o', color='red', markersize=8, label=rf'$\zeta_f = {zeta_threshold:.2f}$', linestyle='None')
plt.xlabel(r'$\zeta$')
plt.ylabel(r'$\theta$')
plt.xlim(0, 4)
plt.legend(fontsize=16, loc='upper right')
plt.title(fr'$k_n = {kn/1e9} \, \mathrm{{GPa/m}}$')
plt.show()
```

```python
overpressure_front = []
x_3dec = P_x_t[kn].iloc[0, 1:].astype(float).to_numpy()
x_3dec_interp = np.linspace(0, 100, 1000)

fig, ax = plt.subplots(figsize=(3.3, 2.7), dpi=300)

for row_index in range(1, P_x_t[kn].shape[0], 10):
    t = float(P_x_t[kn].iloc[row_index, 0])
    P_x_3dec = P_x_t[kn].iloc[row_index, 1:].astype(float).to_numpy()
    P_x_3dec_interp = np.interp(x_3dec_interp, x_3dec, P_x_3dec)
    threshold_indices = np.where(P_x_3dec_interp > overpressure_threshold)[0]
    if len(threshold_indices) > 0:
        index_when_threshold_reached = threshold_indices[-1]
        x_when_threshold_reached = x_3dec_interp[index_when_threshold_reached]
        overpressure_front.append((t, x_when_threshold_reached))

t_front_3dec, x_front_3dec = zip(*overpressure_front)
ax.plot(t_front_3dec, x_front_3dec, linestyle='', label='3DEC', marker='o', markersize=4, color='black')
times = np.linspace(0, float(P_x_t[kn].iloc[-1, 0]), 1000)
D = (kn * w0_[kn] ** 3) / (12 * MU)
ax.plot(times, zeta_threshold * np.sqrt(D * times), label=rf'''Semi-analytical 
        $\zeta_{{\Delta P}} \sqrt{{D t}}$''', color='blue')

ax.set(
    xlabel=r'Time [s], $t$',
    ylabel=r'Distance [m], $x$',
)

ax.legend(loc='upper left', fontsize=8)
# ax.text(1.8, 46, fr'$D = {D:.1f}$ m$^2$/s')
# plt.title(fr'''0.5 MPa overpressure front in a
#           fracture with $k_n = {kn/1e9}$ GPa''')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(3.3, 2.7), dpi=300)

for kn in KN_VALUES:
    if kn != 200e9:
        zeta, theta = sds.solve_n3(wi / w0_[kn])
    else:
        zeta, theta = sds.solve_n1(wi / w0_[kn])

    overpressure_threshold = 0.1e6
    zeta_interp = np.linspace(0, 4, 3000)
    theta_interp = np.interp(zeta_interp, zeta, theta)
    plt.plot(zeta_interp, theta_interp, '.', label='Analytical solution', color='black')
    theta_threshold = (overpressure_threshold / kn + wi) / w0_[kn]
    threshold_indices = np.where(theta_interp > theta_threshold)[0]
    if len(threshold_indices) > 0:
        index_threshold = threshold_indices[-1]
        zeta_threshold = zeta_interp[index_threshold]
        # plt.plot(zeta_threshold, theta_threshold, marker='o', color='red', markersize=5)

    overpressure_front = []
    x_3dec = P_x_t[kn].iloc[0, 1:].astype(float).to_numpy()
    x_3dec_interp = np.linspace(0, 100, 1000)

    for row_index in range(1, P_x_t[kn].shape[0], 10):
        t = float(P_x_t[kn].iloc[row_index, 0])
        P_x_3dec = P_x_t[kn].iloc[row_index, 1:].astype(float).to_numpy()
        P_x_3dec_interp = np.interp(x_3dec_interp, x_3dec, P_x_3dec)
        threshold_indices = np.where(P_x_3dec_interp > overpressure_threshold)[0]
        if len(threshold_indices) > 0:
            index_when_threshold_reached = threshold_indices[-1]
            x_when_threshold_reached = x_3dec_interp[index_when_threshold_reached]
            overpressure_front.append((t, x_when_threshold_reached))

    t_front_3dec, x_front_3dec = zip(*overpressure_front)
    t_front_3dec = np.array(t_front_3dec)
    x_front_3dec = np.array(x_front_3dec)
    mask = x_front_3dec < 100
    ax.plot(t_front_3dec[mask], x_front_3dec[mask], linestyle='', label='3DEC', marker='o', markersize=4, color='black')
    times = np.linspace(0, t_front_3dec[mask][-1], 100)
    if kn != 200e9:
        D = (kn * w0_[kn] ** 3) / (12 * MU)
    else:
        # For kn = 200 GPa, use the same D as in the first plot
        D = (60.5e9 * w0_[kn] ** 3) / (12 * MU)
    ax.plot(times, zeta_threshold * np.sqrt(D * times), label=rf'''Semi-analytical 
            $\zeta_{{\Delta P}} \sqrt{{D t}}$''', color='blue')

    ax.set(
        xlabel=r'Time [s], $t$',
        ylabel=r'Distance [m], $x$',
    )

    # ax.legend(loc='upper left', fontsize=8)
    print(D)
    # ax.text(1.8, 46, fr'$D = {D:.1f}$ m$^2$/s')
# plt.title(fr'''0.5 MPa overpressure front in a
#           fracture with $k_n = {kn/1e9}$ GPa''')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# KN_VALUES = [50e9, 200e9, 500e9]   # example list of k_ns
colors  = plt.cm.tab10(np.linspace(0,1,len(KN_VALUES)))
markers = ['o','s','^','d','v','x','+','*','p','h']

fig, ax = plt.subplots(figsize=(4,3), dpi=300)

overpressure_threshold = 0.1e6

for i, kn in enumerate(KN_VALUES):
    color = colors[i]
    m_data = markers[i%len(markers)]    # marker for 3DEC data
    m_line = markers[(i+1)%len(markers)]  # marker for semi-analytical

    #  solve similarity (n=3 except special case) 
    if kn != 200e9:
        zeta, theta = sds.solve_n3(wi / w0_[kn])
    else:
        zeta, theta = sds.solve_n1(wi / w0_[kn])

    #  find zeta_threshold from analytical solution 
    zeta_interp  = np.linspace(0, 4, 3000)
    theta_interp = np.interp(zeta_interp, zeta, theta)
    theta_th = (overpressure_threshold/kn + wi) / w0_[kn]
    idx_th   = np.where(theta_interp>theta_th)[0]
    if len(idx_th)==0:
        continue
    zeta_th = zeta_interp[idx_th[-1]]

    #  extract 3DEC front points 
    overpres = []
    x3    = P_x_t[kn].iloc[0,1:].astype(float).to_numpy()
    x3int = np.linspace(0,100,1000)
    for row in range(1, P_x_t[kn].shape[0], 10):
        t = float(P_x_t[kn].iloc[row,0])
        P  = P_x_t[kn].iloc[row,1:].astype(float).to_numpy()
        Pint = np.interp(x3int, x3, P)
        idx = np.where(Pint>overpressure_threshold)[0]
        if len(idx)>0:
            overpres.append((t, x3int[idx[-1]]))
    if not overpres:
        continue
    t3, x3front = zip(*overpres)
    t3 = np.array(t3); x3front = np.array(x3front)
    mask = x3front<100

    #  plot 3DEC data 
    ax.plot(t3[mask], x3front[mask],
            linestyle='',
            marker=m_data,
            color=color,
            # markevery=20,  # only show a marker every 20 points
            markersize=4,
            alpha=0.8,
            label=f'3DEC, $k_n={kn/1e9:.0f}\\,$GPa')

    #  semianalytical curve 
    if kn!=200e9:
        D = (kn * w0_[kn]**3)/(12*MU)
    else:
        D = (60.5e9 * w0_[kn]**3)/(12*MU)
    t_fit = np.linspace(0, t3[mask][-1], 200)
    x_fit = zeta_th * np.sqrt(D*t_fit)
    ax.plot(t_fit, x_fit,
            linestyle='-',
            color=color,
            linewidth=1.2,
            alpha=0.9,
            label=f'Semi, $k_n={kn/1e9:.0f}\\,$GPa')

# finalize
ax.set(xlabel=r'Time [s]', ylabel=r'Front position [m]')
ax.legend(fontsize=7, ncol=3)
ax.grid(linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
markers = {30e9: '^', 2000e9: 'o'}
skip = {30e9: 1000, 2000e9: 30}
colors = {30e9: 'tab:blue', 2000e9: 'tab:orange'}

plt.figure(figsize=(3.3, 2.5))

for kn in KN_VALUES:
    if kn == 200e9:
        continue
    zeta, theta = sds.solve_n3(wi / w0_[kn])
    overpressure_threshold = 0.1e6
    zeta_interp = np.linspace(0, 4, 3000)
    theta_interp = np.interp(zeta_interp, zeta, theta)
    theta_threshold = (overpressure_threshold / kn + wi) / w0_[kn]
    threshold_indices = np.where(theta_interp > theta_threshold)[0]
    if len(threshold_indices) > 0:
        index_threshold = threshold_indices[-1]
        zeta_threshold = zeta_interp[index_threshold]
        print(f'For kn = {kn/1e9} GPa, zeta_threshold = {zeta_threshold:.3f}')
    else:
        print(f'For kn = {kn/1e9} GPa, no threshold found')
    overpressure_front = []
    x_3dec = P_x_t[kn].iloc[0, 1:].astype(float).to_numpy()
    x_3dec_interp = np.linspace(0, 100, 1000)
    for row_index in range(1, P_x_t[kn].shape[0], skip[kn]):
        t = float(P_x_t[kn].iloc[row_index, 0])
        P_x_3dec = P_x_t[kn].iloc[row_index, 1:].astype(float).to_numpy()
        P_x_3dec_interp = np.interp(x_3dec_interp, x_3dec, P_x_3dec)
        threshold_indices = np.where(P_x_3dec_interp > overpressure_threshold)[0]
        if len(threshold_indices) > 0:
            index_when_threshold_reached = threshold_indices[-1]
            x_when_threshold_reached = x_3dec_interp[index_when_threshold_reached]
            overpressure_front.append((t, x_when_threshold_reached))
    t_front_3dec, x_front_3dec = zip(*overpressure_front)
    t_front_3dec = np.array(t_front_3dec)
    x_front_3dec = np.array(x_front_3dec)
    mask = x_front_3dec < 100
    times = t_front_3dec[mask]
    plt.plot(t_front_3dec[mask], x_front_3dec[mask], linestyle='', marker=markers[kn], markersize=4, label=fr'3DEC, $k_n =$ {kn/1e9} GPa', color=colors[kn])
    D = (kn * w0_[kn] ** 3) / (12 * MU)
    print(f'For kn = {kn/1e9} GPa, D = {D:.2f} m^2/s')
    t_fit = np.linspace(0, times[-1], 200)
    x_fit = zeta_threshold * np.sqrt(D * t_fit)
    plt.plot(t_fit, x_fit, label=fr'Semi, $k_n =$ {kn/1e9} GPa', color=colors[kn])

plt.xlabel(r'Time [s], $t$')
plt.ylabel(r'Front position [m], $x_f$')
plt.ylim(-5, 103)
plt.legend(fontsize=8, frameon=False)
plt.grid(linestyle='--', alpha=0.2)
plt.tight_layout()
plt.savefig('1-bar-overpressure-fronts.svg', bbox_inches='tight')
plt.show()
```
