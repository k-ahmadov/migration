```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
results = np.load('pme_results.npz')
u_hat = results['u_hat']
Nx = int(results['Nx'])
Nt = int(results['Nt'])
ui_hat = results['ui_hat']
T_hat = results['T_hat']
```

```python
x_hat = np.linspace(0, 1, Nx)
L = 200
kn = 50e9
mu = 1e-3
a = kn / (12*mu)
ui = 1e-4
q0 = 1e-3
U_char = np.power(L*q0/a, 1/4)
T_char = L**2/(a*U_char**3)
T = T_hat * T_char
t = np.linspace(0, T, Nt)
u = u_hat * U_char
P = kn * (u - ui)
x = x_hat * L
```

```python
Pc = 1e5

ge = (P >= Pc)                 # (Nt, Nx)
any_ge = ge.any(axis=1)        # any cell ≥ Pc
all_gt = (P > Pc).all(axis=1)  # all cells > Pc
mask = any_ge & ~all_gt        # rows that cross Pc

# rightmost index j with P[k,j] ≥ Pc, but only for rows with a crossing
J_from_right = ge[mask, ::-1].argmax(axis=1)          # distance from the right edge
idx_front = P.shape[1] - 1 - J_from_right             # column indices of the front
idx_times = np.flatnonzero(mask)                      # time indices with a crossing
t_front = t[idx_times]
x_front = x[idx_front]
```

```python
def power_law(t, a, b):
    return a*t**b
```

```python
popt, pcov = curve_fit(power_law, t_front, x_front)
plt.plot(t_front, x_front, '.')
plt.plot(t_front, power_law(t_front, *popt))
```
